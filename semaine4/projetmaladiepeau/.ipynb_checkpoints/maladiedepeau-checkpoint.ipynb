```python
from fastai import *
from fastai.vision import *
```

```python
#pip install google_images_download # à décommenter la ligne lors de la premiére installation
from google_images_download import google_images_download

arguments = {}
# choix des catégories à télécharger sans espace entre les virgules exemple les 4 saisons
arguments["keywords"] = "psoriasis,acné,vitiligo,varicelle"
arguments["limit"] = 100
arguments["print_urls"] = False
arguments["format"] = "jpg" # ne pas changer !!
arguments["color_type"] = "full-color"
#arguments["size"] = ">800*600"
#arguments["aspect_ratio"] = "square"
#arguments["type"] = "photo"
arguments["output_directory"] = os.getcwd()+'/data/'
response = google_images_download.googleimagesdownload()
paths = response.download(arguments)
path = Path(arguments["output_directory"])

import os

# Renommage des images téléchargées et skip du dossier model propre à fast ai.
for folder in path.ls():
    i = 1
    label = str(folder).split('/')[-1]
    if label != 'models':
        for photo in folder.ls():
            new_name = f'{str(folder)}/{label}_{i}.jpg'
            os.rename(photo, new_name)
            i = i + 1

# Définition du path et suppression des fichiers de sauvegardes autos susceptibles d'occasionner 
# des bugs si vous stockez vos classes à la racine du projet .
%cd '{path}'
!find '.' -name '*.ipynb_checkpoints' -exec rm -r {} +

# Suppression des fichiers récalcitrants.
for c in arguments["keywords"].split(','):
    print(c)
    path_to_class_folder = arguments["output_directory"] + "/" + c
    verify_images(path_to_class_folder, delete=True, img_format=f'{c} %d')

# Suppression des espaces dans les noms de fichiers.
def replace(parent):
    for path, folders, files in os.walk(parent):
        for f in files:
            os.rename(os.path.join(path, f), os.path.join(path, f.replace(' ', '_')))
replace(arguments["output_directory"])

path.ls()
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train='.', 
                                  valid_pct=0.2, 
                                  ds_tfms=get_transforms(),
                                  size=224).normalize(imagenet_stats)
data.show_batch(rows=3, figsize=(9,10))
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
#on précise que l'on travaille bien sur GPU et non CPU

defaults.device = torch.device('cuda')
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('maladiepeau_34')
```

```python
#nettoyage du dataset

import sys
!conda install --yes --prefix {sys.prefix} ipywidgets
from fastai.widgets import *
```

```python
#DatasetFormatter permet de retourner le dataset avec les données sous un bon format et correctement indexé.
#from_toplosses permet de prendre les images pour lesquelles le modèle se trompe le plus.

ds, idxs = DatasetFormatter().from_toplosses(learn)
```

```python
ImageCleaner(ds, idxs, path)
```

```python
#On utilise pandas pour lire le fichier cleaned.csv et on affiche avec head les 5 premières données avec le nom des fichiers et leurs classes.
df = pd.read_csv(path/'cleaned.csv', header='infer')
df.head()
```

```python
#On vérifie que les noms de fichiers ne sont pas corrompus.
df[(df['name'].apply(lambda x: len(x)<5))]
```

```python
#On réaffecte un random seed de 8.
#On recrée l'ImageDataBunch avec notre nouveau dataset nettoyé.

np.random.seed(8)
db = (ImageList.from_df(df, path)
     .split_by_rand_pct(0.2)
     .label_from_df()
     .transform(get_transforms(), size=224)
     .databunch(bs=9)).normalize(imagenet_stats)
```

```python
#voici le dataset non nettoyé
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
#voici le dataset nettoyé
db.classes, db.c, len(db.train_ds), len(db.valid_ds)
```

```python
db.show_batch()
```

```python
learn = cnn_learner(db, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('maladiedepeau34-2')
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(1)
```

```python
learn.load('maladiedepeau34-2')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, max_lr=slice(1e-04, 1e-03))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(14,9))
```

```python
interp.plot_confusion_matrix()
```

```python
interp.most_confused(min_val=1)
```
